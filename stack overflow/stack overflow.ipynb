```python
c_max = df['CompTotal'].max()
```

```python
df['CompTotal'].value_counts()[c_max]
```

```python
upper_bound = 4.000000e+05
df['CompTotal'] = np.where(df['CompTotal'] > upper_bound, df['CompTotal'].median(), df['CompTotal'])
```

```python
df['CompTotal'].max()
```

```python
df['WorkExp'].value_counts()
```

```python
plt.scatter(x = df['WorkExp'], y=df['CompTotal'])
```

```python
bins = [1, 3, 5, 10, 20, 35, 50]
labels = ['1-3 years', '3-5 years', '5-10 years', '10-20 years', '20-35 years', '35-50 years']
df['Exp Grouped'] = pd.cut(df['WorkExp'], bins=bins, labels=labels, right=False)
```

```python
sns.barplot(data=df, x='Exp Grouped', y='CompTotal')
plt.show()
```
